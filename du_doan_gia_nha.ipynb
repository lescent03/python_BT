```python
import pandas as pd
```

```python
# load dữ liệu
train = pd.read_csv("./house-prices/train.csv")
test_data = pd.read_csv("./house-prices/test.csv")
test_labels = pd.read_csv("./house-prices/sample_submission.csv")
```

```python
train.head()
#test_data.head()
#test_labels.head()
```

```python
X = ["MSSubClass", "LotArea", "LotShape", "LandContour", "LotConfig", "Neighborhood", "OverallQual", "OverallCond",
     "YearBuilt", "YearRemodAdd", "RoofStyle", "Exterior1st", "Exterior2nd", "MasVnrArea", "Foundation", "BsmtFinType1",
     "BsmtFinSF1", "BsmtUnfSF", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "GrLivArea", "FullBath", "BedroomAbvGr", "KitchenAbvGr",
     "TotRmsAbvGrd", "Fireplaces", "GarageType", "GarageCars", "GarageArea", "WoodDeckSF", "OpenPorchSF","YrSold"]

train_data = train[X]
train_labels = train[["SalePrice"]]
test_data = test_data[X]
test_labels = test_labels[["SalePrice"]]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Mã hoá
Y = ["LotShape", "LandContour", "LotConfig", "Neighborhood", "RoofStyle", "Exterior1st","Exterior2nd", "Foundation",
     "BsmtFinType1", "GarageType"]

label_encoder = LabelEncoder()
for col in Y:
    train_data.loc[:, col] = label_encoder.fit_transform(train_data[col])
    test_data.loc[:, col] = label_encoder.fit_transform(test_data[col])

train_data = train_data.fillna(0)
test_data = test_data.fillna(0)
```

```python
train_data.head()
#print(train_data.count())
#test_labels.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# huấn luyện mô hình
reg = LinearRegression()
reg.fit(train_data, train_labels)
```

```python
# dự đoán
y_preTrain = reg.predict(train_data)
y_preTest = reg.predict(test_data)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# đánh giá
print("MSE train: ", mean_squared_error(train_labels, y_preTrain))
print("MSE test: ", mean_squared_error(test_labels, y_preTest))
```

```python
import math
```

```python
print("RMSE train: ", math.sqrt(mean_squared_error(train_labels, y_preTrain)))
print("RMSE test: ", math.sqrt(mean_squared_error(test_labels, y_preTest)))
```

```python
from sklearn.metrics import r2_score
```

```python
print("r2 train: ", r2_score(train_labels, y_preTrain))
print("r2 test: ", r2_score(test_labels, y_preTest))
```
